### The stack() and unstack() methods

1) __stack()__: The stack() method reshapes the DataFrame by passing a level of indexing from columns to rows. As a result, we will have a DataFrame with an additional indexing level in the rows and one less than in the columns.

2) __unstack()__: The unstack() method does exactly the reverse – reshaping the DataFrame by transferring a level of indexing from rows to columns.

```python
df = pd.DataFrame(np.arange(1,10).reshape(3,3),
                  index=['a','b','c'],
                  columns=['A','B','C'])
print(df,"\n")

dfs = df.stack()
print(dfs)
# this reduces the df to a series
```

```python
# you can revert this with unstack()
print(dfs.unstack())
```

### Pivoting
### applies a reindex and aggregation of the columns

```python
df = pd.DataFrame([['red','x',1,1],
['red','x',0,1],
['black','y',0,2],
['red','y',1,0],
['red','y',2,0],
['black','x',2,1]],
columns=['A','B','C','D'])
print(df)
```

The pivoting operation is performed in Pandas through the **pivot_table()** method. This method accepts several optional parameters such as columns *which converts the values present in the column to indexes for the columns of the pivot* (aggregations of common types). The *index parameter* converts the values in the column to indexes for the rows of the pivot (aggregation of common types). The index labels of the original DataFrame are not considered. For each label aggregation, the corresponding values in the starting DataFrame are aggregated using the function specified with the optional **aggfunc** parameter, as shown as follows:""

```python
table = pd.pivot_table(df, values=['C','D'],
index=['A'],
columns=['B'],
aggfunc=np.sum)
print(table)
```

You can extend the pivot table by applying different aggregating functions for each column of values. The **aggfunc** parameter can, in fact, accept a dict, in which the keys are the columns while the values are the aggregating functions. If we want to
apply more functions to the same column, we can insert a list as a value, as shown as follows:

```python
table = pd.pivot_table(df, values=['C','D'],
index=['A'],
columns=['B'],
aggfunc={'C': np.sum,
'D': [min, max, np.mean]})
print(table)
```

### Applying functions with the apply () method

**apply()** method takes a function and, by default, applies it to the DataFrame column by column, considering their values as input. The result will then depend on what the function is programmed to do. If the function
used is aggregating, that is, it takes an input list and returns a single value, as a result, we will get a Series, with each element corresponding to a column. Otherwise, it will operate element by element, and in this case, we
will obtain as a result a DataFrame of the same size as the original one.

```python
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
index=['a','b','c'],
columns=['A','B','C'])
print(df)

df.apply(np.sum)
```

```python
# line-by-line would be done as follows
df.apply(np.sum, axis=1)
```

```python
# a function such as np.sqrt, however, evaluates element-wise 
df2 = df.apply(np.sqrt)
print(df2)
```

```python
#consider how you can apply customized functions via apply()
def double_up(x):
    return x * 2
df2 = df.apply(double_up)
print(df2)
```

```python
# lambda is even better because you can define the function inline
print(df.apply(lambda x: x.max() - x.min()))

# or the previous example

print(df.apply(lambda x: x*2))

# set the axis if need be

print(df.apply(lambda x: x.max() - x.min(), axis=1))
```

### Applications of functions with the pipe() method
**pipe()** method allows you to apply functions on the DataFrame where **apply()** is unviable

```python
df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],
index=['a','b','c'],
columns=['A','B','C'])
print(df)

#consider if we have already defined a max function
def maximus(x):
    return np.max(np.max(x))

#to pass it to the df, just use the pipe() method
print(df.pipe(maximus),"\n")

#and here is the difference with apply() (it's row/column wise)
df.apply(maximus)
```

#### Chaining pipe() and apply() methods
Chaining will allow you to apply functions to incoming objects

```python
print(df.pipe(np.max))
print(df.pipe(np.max).pipe(np.max))
```

```python
# more advanced examples
def dubble_up(x):
    return x * 2
print(df.pipe(dubble_up).pipe(np.mean).pipe(lambda x: x-2))

print(df.apply(dubble_up).apply(np.mean).apply(lambda x: x-2))
```

### Transforming

```python
# The transform() method applies transformations on the starting DataFrame, by applying functions
df = pd.DataFrame(np.arange(1,10).reshape(3,3),
index=['a','b','c'],
columns=['A','B','C'])
print(df,"\n")

def double(x):
    return x*2
df2 = df.transform(double)
print(df2,"\n")

## see how this is identical to apply() though
df2 = df.apply(double)
print(df2)
```

```python
#The transform() method can work with multiple functions at the same time.
## These can be passed either as a list or as a dict, which the apply() method cannot do
multidf = df.transform([np.sqrt, double])
print(multidf)
```

```python
# Reducing a MultiIndex DataFrame to a DataFrame
## this gives the df from of the double function
### IndexSlice, necessary to make the selections on several levels, and the droplevel() method, necessary to eliminate the level of ndexes from the columns
idx = pd.IndexSlice
dfn = multidf.loc[idx[:], idx[:, 'double']]
dfn.columns = dfn.columns.droplevel(1)
print(dfn,"\n")
df2 = df.transform({'A': np.sqrt, 'B': np.double})
print(df2)
```

```python
# transform does not work with aggregation
    #print(df.transform(np.sum))
# but apply() can
print(df.apply(sum))
```

```python
#Transform() works with a single column at a time
def adding(x):
    return x[0] + x[1]
df.apply(adding, axis=1)

"""
Transforming means transforming one DataFrame into another,manipulating the data,
making it evolve into something different. They areimposed to avoid that;
sometimes even unconsciously, the structure of the DataFrame being analyzed can be varied in some way.
"""
```

### Grouping

```python
df = pd.DataFrame(np.random.randint(1,5,18).reshape(6,3),
                  index=['a','b','c','a','b','c'],
                  columns=['A','B','C'])
print(df.sort_values(by=['A']))
```

#### splitting


```python
# representing max values for each row of A's groups
dfg = df.groupby(by='A').max()
print(dfg)
```

```python
#The groupby() method for indexes

tuples = [('x', 'a'), ('x', 'b'), ('y', 'a'), ('y','b')]
index = pd.MultiIndex.from_tuples(tuples,
names=['first','second'])
tuples = [('X', 'A'), ('X', 'B'), ('Y', 'A'), ('Y','B')]
columns = pd.MultiIndex.from_tuples(tuples, names=['high','low'])
dfm = pd.DataFrame(np.arange(1,17).reshape(4,4), index = index, columns = columns)
print(dfm)
```

```python
# then apply level 0 grouping as follows
df2 = dfm.groupby(level=0).max()
print(df2)

"""In this case, the rows with the same label at level 0 of the Index of row (that is, x and y)
are grouped, and within the group, the greater value is chosen for each column.
Information regarding the second-level labels of the index is lost.
"""
```

```python
# level 1 grouping yields this
df2 = dfm.groupby(level=1).max()
print(df2)
```

```python
# further arguments yield the following:
df2 = dfm.groupby(level=0, axis=1).max()
print(df2,"\n","="*20)

df2 = dfm.groupby(level=1, axis=1).max()
print(df2)
```

#### Grouping with transformation

```python
gdf = df.groupby('A').transform(np.sum)
print(gdf)

"""
With the transformation, the values contained in column A have been lost.
They only served to subdivide into computational groups. The results
obtained by the group were then assigned to every single row according to
their belonging group. In fact, there are lines with identical results.
"""
```

#### Grouping with aggregation

```python
gdf = df.groupby('A').agg(np.sum)
print(gdf)

"""
Although the same np.sum() function has been applied, the results obtained
are significantly different between transformation and aggregation. In fact,
the aggregation necessarily produces a smaller result. The number of rows
will now correspond to the existing groups, while for the other columns, the
aggregating function replaces the different values with a single value that is the result.
Inversely, from the case of the transformation, it is the index labels of the
original DataFrame that are no longer present, since in the new system, they
have lost their meaning, while the values of column A have become the new labels.
"""
```

#### Apply() with groupby()

```python
# this yields 1 result per group
gdf = df.groupby('A').apply(np.sum)
print(gdf)

"""
As far as column A is concerned, we
have the same behaviour as the aggregation, with the conversion of its
values into index labels. We, therefore, have an aggregation that has led to
the reduction of the DataFrame to only 4 lines, each for each group. But at
the same time, there are some differences, such as column A which still
remained as such in the DataFrame, and whose values are the result of the aggregation (sum) such as columns B and C
"""
```

### Pie Charts

```python
fig, ax = plt.subplots()
ax.pie(y)
#this controls transparency
fig.patch.set_alpha(0)
plt.show()
```

* get_cmap() specifies base color
* np.linspace(), can create as many shades of a color as there are elements (slices of cake) to represent.
* wedgeprops parameter has further configurable parameters such as linewidth and edgecolor which are the width and color of the dividing line between the slices.
* associate a corresponding numerical value to the corresponding slice with **autopct**

```python
fig, ax = plt.subplots()
colors = plt.get_cmap('Oranges')(np.linspace(0.2, 0.7, len(x)))
ax.pie(y,radius=1,colors=colors,wedgeprops={"linewidth": 1,"edgecolor": "lightgreen"},autopct="%.2f")
ax.legend(x, bbox_to_anchor=(0, 1), loc='best', ncol=1)
fig.patch.set_alpha(0)
plt.show()
```
